### Plotting multiple lines

````python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
````

````python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")

plt.title("Crop yields in Kanto")
plt.legend(["Apples", "Oranges"])
````

### Line markers

````python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")

plt.title("Crop yields in Kanto")
plt.legend(["Apples", "Oranges"])
````

### Styling lines and markers

The `plt.plot` function supports many arguments for styling lines and markers:

- `color` or `c`: Set the color of the line ([supported colors](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` or `ls`: Choose between a solid or dashed line
- `linewidth` or `lw`: Set the width of a line
- `markersize` or `ms`: Set the size of markers
- `markeredgecolor` or `mec`: Set the edge color for markers
- `markeredgewidth` or `mew`: Set the edge width for markers
- `markerfacecolor` or `mfc`: Set the fill color for markers
- `alpha`: Opacity of the plot


Check out the documentation for `plt.plot` to learn more: [https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) .

````python
plt.plot(years, apples, marker='o', c='g', ls='-', lw=2, mec='b')
plt.plot(years, oranges, marker='x', c='r', ls='--', lw=3, mec = 'black')

plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title("Crop yields in Kanto")
plt.legend(["Apples", "Oranges"])
````

- Shorthand for specifying the line style, marker, and line color
- It can be given as third arg in ```plt.plot```
- ```fmt = '[marker][line][color]'```

````python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title("Crop yields in Kanto")
plt.legend(["Apples", "Oranges"]);
````

# Scatter plot

1. The values of 2 variables are plotted as points on a 2-dim grid.
2. Additionally, you can use a 3rd variable to determine the size or color of the points

````python
flowers_df = sns.load_dataset("iris")
````

````python
flowers_df
````

````python
flowers_df.species.unique()
````

````python
# relationship between sepal_length and sepal_width etc

plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
````

````python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
````

- Now we see all the points but not any relatioinships
- We can maybe say vaguely ~3 cluster of points present
- But not very clear from above figure

### Adding Hues
- We can color the dots using species as hue

````python
plt.figure(figsize=(8,4))
plt.title("Sepal dimensions")
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=50);
````

### Plotting using pandas dataframe
- Seaborn have support for pandas df
- so, instead of passing df columns as series, we can pass column names and additional ```data``` argument for df

````python
plt.figure(figsize=(8,4))
plt.title("Sepal dimensions")

sns.scatterplot(x='sepal_length',
                y='sepal_width',
                hue=flowers_df.species,
                s=50,
               data = flowers_df);
````

# Histogram
- represents distribution of data using bins

````python
flowers_df.sepal_width
````

````python
flowers_df.describe()
````

````python
plt.title("Sepal width distribution")
plt.hist(flowers_df.sepal_width);
````

````python
plt.title("Sepal width distribution")
plt.hist(flowers_df.sepal_width, bins=5);
````

````python
import numpy as np

plt.title("Sepal width distribution")
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
````

````python
plt.title("Sepal width distribution")
plt.hist(flowers_df.sepal_width, bins=[1,3,4,4.5]); # unequal bins
````

### Multiple histograms

````python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
````

````python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25));
plt.legend(['Setosa', 'Versicolor'])
````

#### Stacking histograms

````python
sns.set_style("darkgrid")
````

````python
plt.title("Distribution of Sepal Widths")
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
        stacked=True,
        bins=np.arange(2,5,0.25));
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
````

#### Bar plot with averages
- Sometimes we may not just want to show a single metric on BAR chart
- But, something like average of few columns or rows etc

````python
# tips dataset: tips given by customers
tips_df = sns.load_dataset('tips')
````

````python
tips_df.head()
````

````python
# what does the average bill look like on a average Sunday?
bill_avg_df = tips_df.groupby('day')[['total_bill']].mean() # avg total bills on each dayshat
````

````python
plt.bar(bill_avg_df.index, bill_avg_df.total_bill)
````

- The 2 operations shown above are pretty common, say next you'd want to see avg tips across gender/time etc
- Hence, ```seaborn``` inbiult support to do this in ```sns.barplot```

````python
sns.barplot(x='day', y='total_bill', data=tips_df);
````

````python
sns.barplot(x='day', y='tip', data=tips_df, hue='sex');
````

````python
sns.barplot(x='tip', y='day', data=tips_df, hue='sex');
````

- to use a heatmap

````python
flights_df
````

````python
plt.title("No. of passengers")
sns.heatmap(flights_df);
````

````python
plt.title("No. of passengers (1000s)")
sns.heatmap(flights_df, annot=True, cmap="Blues", fmt='d');
````
