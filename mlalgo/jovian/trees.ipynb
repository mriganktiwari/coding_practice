````python
import opendatasets as od
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python
raw_df = pd.read_csv('weather-dataset-rattle-package/weatherAUS.csv')
# raw_df.head()
````

````python
# raw_df.info()
````

````python
raw_df.dropna(subset=['RainTomorrow'], inplace=True)
# raw_df.info()
````

````python
year = pd.to_datetime(raw_df.Date).dt.year

train_df = raw_df[year < 2015]
val_df = raw_df[year == 2015]
test_df = raw_df[year > 2015]

train_df.shape, val_df.shape, test_df.shape
````

````python
input_cols = list(train_df.columns)[1:-1]
target_col = 'RainTomorrow'

# input_cols
````

````python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_col].copy()

val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_col].copy()

test_inputs = test_df[input_cols].copy()
test_targets = test_df[target_col].copy()
````

````python
# train_df.select_dtypes(include=np.number).columns.tolist()
train_inputs.select_dtypes(include='object').columns.tolist()
````

````python
numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols = train_inputs.select_dtypes(include='object').columns.tolist()
````

````python
raw_df[categorical_cols].nunique()
````

## How do we figure out best value for ```max_depth``` ?

````python
def max_depth_error(md):
    model = DecisionTreeClassifier(max_depth=md, random_state=42)
    model.fit(X_train, train_targets)
    train_error = 1 - model.score(X_train, train_targets)
    val_error = 1 - model.score(X_val, val_targets)
    return {'Max depth':md, 'Training error':train_error, 'Validation error':val_error}
````

````python
# for i in range(1,21):
#     print(max_depth_error(i))

error_df = pd.DataFrame([max_depth_error(md) for md in range(1,21)])
error_df
````

````python
plt.figure()
plt.plot(error_df['Max depth'], error_df['Training error'])
plt.plot(error_df['Max depth'], error_df['Validation error'])
plt.title("Training vs Validation error")
plt.xticks(range(1,21))
plt.xlabel("Max depth")
plt.ylabel("Prediction error (1-accuracy)")
plt.legend(["training", "validation"]);
````

- At ```max_depth``` of 7, we can get best validation error

````python
model = DecisionTreeClassifier(max_depth=7, random_state=42)
model.fit(X_train, train_targets)

model.score(X_train, train_targets), model.score(X_val, val_targets)
````

- Another way to control complexity of DT is ```max_leaf_node```

````python
model = DecisionTreeClassifier(max_leaf_nodes=128, random_state=42)
model.fit(X_train, train_targets)

model.score(X_train, train_targets), model.score(X_val, val_targets)
````

# Random Forests

- Tuning parameters in DTs like max_depth or max_leaf_depth can lead to some improvements
- But much effective technique is to combine results of several DTs
- There is some kind of randomization involved in how these several trees are created
- Different trees will make diff kind of errors
- And on combining them will cancel out many of those errors.

````python
from sklearn.ensemble import RandomForestClassifier
````

````python
model = RandomForestClassifier(n_jobs=-1, random_state=42)
````

````python
%%time
model.fit(X_train, train_targets)
````

````python
model.score(X_train, train_targets), model.score(X_val, val_targets)
````

````python
train_probs = model.predict_proba(X_train)
train_probs
````

````python
len(model.estimators_)
````

## Hyper parameter tuning with RFs

````python
base_model = RandomForestClassifier(n_jobs=-1, random_state=42).fit(X_train, train_targets)

base_train_acc = base_model.score(X_train, train_targets)
base_val_acc = base_model.score(X_val, val_targets)
````

````python
base_accs = base_train_acc, base_val_acc
base_accs
````

### ```n_estimators```

````python
model = RandomForestClassifier(n_estimators=10, n_jobs=-1, random_state=42).fit(X_train, train_targets)
model.score(X_train, train_targets), model.score(X_val, val_targets)
````

````python
model = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=42).fit(X_train, train_targets)
model.score(X_train, train_targets), model.score(X_val, val_targets)
````

````python
def test_params(**params):
    model = RandomForestClassifier(n_jobs=-1, random_state=42, **params).fit(X_train, train_targets)
    return model.score(X_train, train_targets), model.score(X_val, val_targets)
````

## ```max_depth``` / ```max_leaf_node``` for RFs

````python
test_params(max_depth=5)
````

````python
test_params(max_depth=26)
````

````python
test_params(max_leaf_nodes=2**5)
````

````python
test_params(max_leaf_nodes=2**20)
````
